```python
def genLaTex(ipa):
    latex = ''
    for g in ipa:
        try:
            if 'not' in ipa_dict[g]:
                print('not found\t "' + g + '"')
                latex += ' [not found] '
            else:
                latex += ipa_dict[g]
        except KeyError as ke:
            print(ipa, 'key error\t "' + g + '"',"\t",len(g))
    return latex
```

# Testing top 3000 words in English contributed by Longman

```python
with open('word-ipa.tsv', 'r') as file:
    latexLine = ''
    for i,line in enumerate(file):
        word, ipa = line.strip().split('\t')
        latexLine = "\\textbf{ " + word + " }\quad " + genLaTex(ipa) + "\\\\"
        print(latexLine)
```
